## groupby 版本的 FoodFacts

```python
import pandas as pd
```

```python
filename = "./FoodFacts.csv"
csv_obj = pd.read_csv(filename,usecols = ['countries_en', 'additives_n'])
df_obj = csv_obj.dropna()
```

```python
print(df_obj["additives_n"].mean())
```

```python
data = df_obj["additives_n"].groupby(df_obj["countries_en"]).mean()
#print(data)
```

```python
data = data.sort_values(ascending=False)
```

```python
%matplotlib inline
data[:10].plot.bar()
```
